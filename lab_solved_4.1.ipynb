```python
import pandas as pd 

customer_df = pd.read_csv("customer_value_analysis.csv")
customer_df.head()
customer_df.shape
customer_df.info()
```

```python
column_mapping = {col: col.lower().replace(' ', '_') for col in customer_df.columns}
customer_df.rename(columns=column_mapping, inplace=True)
customer_df.head()
```

```python
customer_df['effective_to_date'] = pd.to_datetime(customer_df['effective_to_date'])
```

```python
nan_count = customer_df.isna().sum()
print(nan_count)
```

```python
customer_df.drop('customer', axis=1)

def return_two_df(df):
    # Identify numerical columns
    num_df = df.select_dtypes(include=['int', 'float'])

    # Identify categorical columns
    cat_df = df.select_dtypes(include=['object'])
    
    return num_df, cat_df

num_df, cat_df = return_two_df(customer_df)
```

```python
correlation_matrix = num_df.corr()
print(correlation_matrix)
```

```python

def return_cont_discrete_df(df, threshold=36):
    continuous_columns = []
    discrete_columns = []
    
    for column in df.select_dtypes(include=['int', 'float']):
        unique_values = df[column].nunique()
        
        if unique_values <= threshold:
            discrete_columns.append(column)
        else:
            continuous_columns.append(column)
    
    continuous_df = df[continuous_columns]
    discrete_df = df[discrete_columns]
    
    return continuous_df, discrete_df

continuous_df, discrete_df = return_cont_discrete_df(customer_df)
```
